```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('height_weight.csv')
df.shape
```

```python
df.head()
```

```python
sn.histplot(df['height'], kde=True)
plt.show()
```

```python
sn.histplot(df['weight'], kde=True)
plt.show()
```

```python
weight_Q1 = df['weight'].quantile(0.25)
weight_Q3 = df['weight'].quantile(0.75)
weight_IQR = weight_Q3 - weight_Q1

lower_weight = weight_Q1 - 1.5 * weight_IQR
upper_weight = weight_Q3 + 1.5 * weight_IQR
lower_weight, upper_weight
```

```python
weight_outliers = df[(df['weight'] < lower_weight) | (df['weight'] > upper_weight)]
weight_outliers
```

```python
df.drop(weight_outliers.index, axis='index', inplace=True)
```

```python
height_Q1 = df['height'].quantile(0.25)
height_Q3 = df['height'].quantile(0.75)
height_IQR = height_Q3 - height_Q1

lower_height = height_Q1 - 1.5 * height_IQR
upper_height = height_Q3 + 1.5 * height_IQR
lower_height, upper_height
```

```python
height_outliers = df[(df['height'] < lower_height) | (df['height'] > upper_height)]
height_outliers
```

```python
df.drop(height_outliers.index, axis='index', inplace=True)
```

```python
sn.histplot(df['height'], kde=True)
plt.show()
```

```python
sn.histplot(df['weight'], kde=True)
plt.show()
```
