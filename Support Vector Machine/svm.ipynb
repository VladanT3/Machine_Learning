```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
iris = load_iris()
dir(iris)
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df['target_name'] = df['target'].apply(lambda x: iris.target_names[x])
df.head()
```

```python
df_setosa = df[df['target'] == 0]
df_versicolor = df[df['target'] == 1]
df_virginica = df[df['target'] == 2]
```

```python
plt.scatter(df_setosa['sepal length (cm)'], df_setosa['sepal width (cm)'], marker='.', color='blue')
plt.scatter(df_versicolor['sepal length (cm)'], df_versicolor['sepal width (cm)'], marker='+', color='red')
plt.scatter(df_virginica['sepal length (cm)'], df_virginica['sepal width (cm)'], marker='x', color='green')
plt.title('Sepal width and length')
plt.xlabel('Length')
plt.ylabel('Width')
plt.show()
```

```python
plt.scatter(df_setosa['petal length (cm)'], df_setosa['petal width (cm)'], marker='.', color='blue')
plt.scatter(df_versicolor['petal length (cm)'], df_versicolor['petal width (cm)'], marker='+', color='red')
plt.scatter(df_virginica['petal length (cm)'], df_virginica['petal width (cm)'], marker='x', color='green')
plt.title('Petal width and length')
plt.xlabel('Length')
plt.ylabel('Width')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
model = SVC()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
