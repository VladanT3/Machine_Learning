```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
canada_GDP = pd.read_csv('canada_per_capita_income.csv')
canada_GDP.head()
```

```python
plt.xlabel('Year')
plt.ylabel('per capita income ($)')
plt.scatter(canada_GDP.year, canada_GDP['per capita income (US$)'])
plt.plot(canada_GDP.year, canada_GDP['per capita income (US$)'])
plt.show()
```

```python
lin_reg = linear_model.LinearRegression()
lin_reg.fit(canada_GDP['year'].values.reshape(-1, 1), canada_GDP['per capita income (US$)'])
```

```python
plt.xlabel('Year')
plt.ylabel('per capita income ($)')
plt.scatter(canada_GDP.year, canada_GDP['per capita income (US$)'])
plt.plot(canada_GDP.year, canada_GDP['per capita income (US$)'])
plt.plot(canada_GDP.year, lin_reg.predict(canada_GDP['year'].values.reshape(-1, 1)), color='red')
plt.show()
```
